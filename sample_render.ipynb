```python
import requests
import pandas as pd
```

```python
base_url = "https://reder-test-o5k8.onrender.com"
```

```python
# 1. 利用可能な国リストを取得
countries =requests.get(f"{base_url}/get_countries").json()
df_countries = pd.DataFrame(countries)
df_countries
```

```python
events = requests.get(f"{base_url}/get_events").json()
df_events = pd.DataFrame(events)
df_events
```

```python
# 2. 例：中国（CHN）の地震イベントを取得

events_url = f"{base_url}/get_events"

china_earthquakes = df_events[
    (df_events['country_iso3'] == 'CHN') &
    (df_events['disaster_type'].str.lower() == 'earthquake')
]

china_earthquakes
```

```python
# kmzファイルの取得
```

```python
product_list = []
for _, event in china_earthquakes.iterrows():
    if not event['url']:
        continue
    products_url = f"{base_url}/get_products"
    res = requests.get(products_url, params={"url": event['url']})
    if res.status_code != 200:
        continue
    products = res.json()
    kmz_products = [p for p in products if p.get('file_type') == 'kmz']
    for prod in kmz_products:
        product_list.append({
            "event_name": event['name'],
            "product_title": prod.get('title'),
            "download_url": prod.get('download_url')
        })
```

```python
df_products = pd.DataFrame(product_list)
df_products
```

```python
# 3. 災害タイプで「Earthquake」のみ抽出
target_events = [e for e in events if e['disaster_type'].lower() == 'earthquake']
target_events
```

```python
# 4. 各イベントのProductを取得し、kmzファイルのみ抽出

product_list = []
for event in target_events:
    if not event.get('url'):
        continue
    products_url = f"{base_url}/get_products"
    res = requests.get(products_url, params={"url": event['url']})
    if res.status_code != 200:
        continue
    products = res.json()
    kmz_products = [p for p in products if p.get('file_type') == 'kmz']
    for prod in kmz_products:
        product_list.append({
            "event_name": event['name'],
            "product_title": prod.get('title'),
            "download_url": prod.get('download_url')
        })
```

```python
df_products = pd.DataFrame(product_list)
df_products
```
